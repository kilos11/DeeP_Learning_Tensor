```python
import pandas as pd

Advertise_df = pd.read_csv("/content/drive/MyDrive/IRIS/Advertising_2023.csv")
```

```python
#Used To  Get Consice Summery of The Dataset
Advertise_df.info()
```

```python
#Get Summery Of The Dataset
Advertise_df.describe()
```

```python
#Shape of the Dataframe
Advertise_df.shape
```

```python
#The isnull() method is used to check null values in the dataset
Advertise_df.isnull().sum()
```

```python
#X(features) and y(target)
X = Advertise_df.drop('sales', axis=1)
y = Advertise_df['sales']
```

```python
#Feature Normalization
import keras
from sklearn.preprocessing import MinMaxScaler

normalized_feature = keras.utils.normalize(X.values)
print(normalized_feature)
```

```python
#import train_test_split split the funtion
from sklearn.model_selection import train_test_split

#split the data into training and testing
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
print(X_train.shape, X_test.shape, y_train.shape, y_test.shape)
```

```python
# Import the Keras library
import keras

# Build a Sequential model
model = keras.Sequential()

# Add a Dense layer with 10 units and ReLU activation function
model.add(keras.layers.Dense(10, activation='relu'))

# Add another Dense layer with 9 units and ReLU activation function
model.add(keras.layers.Dense(9, activation='relu'))

# Add a Dense output layer with 1 unit (for regression tasks, as there is no activation specified)
model.add(keras.layers.Dense(1))
```

```python
#Complile Model
model.compile(optimizer='adam',loss = "mse", metrics=['mse'])
```

```python
#Fit The Model
history = model.fit(X_train, y_train,validation_data=(X_test, y_test), epochs=100)
```

USING VISIUALISATION FOR EVALUATION OF THE MODEL


```python
# Import necessary visualization libraries
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline

# Plot the training and validation loss over epochs
plt.plot(history.history['loss'])         # Plot training loss
plt.plot(history.history['val_loss'])     # Plot validation loss
plt.title('Model Loss')                   # Set the title of the plot
plt.ylabel('Loss')                        # Label for the y-axis
plt.xlabel('Epoch')                       # Label for the x-axis
plt.legend(['Train', 'Test'], loc='upper right')  # Add a legend to indicate Train and Test lines in the plot
plt.show()                               # Display the plot
```

```python
#Predictions
test_predictions = model.predict(X_test).flatten()
print(test_predictions)
```

```python
# Create a DataFrame 'true_predictions' to display true values and corresponding predictions
true_predictions = pd.DataFrame(list(zip(y_test, test_predictions)), columns=['True Values', 'Predictions'])

# Print the first few rows of the DataFrame to inspect the true values and predictions
print(true_predictions.head())
```

```python
# Import necessary libraries
import numpy as np
from sklearn.metrics import mean_squared_error

# Use the trained model to predict on the training set
pred_train = model.predict(X_train)

# Calculate and print the root mean squared error (RMSE) between the true values (y_train) and predictions on the training set
print(np.sqrt(mean_squared_error(y_train, pred_train)))

# Use the trained model to predict on the test set
pred = model.predict(X_test)

# Calculate and print the root mean squared error (RMSE) between the true values (y_test) and predictions on the test set
print(np.sqrt(mean_squared_error(y_test, pred)))
```
